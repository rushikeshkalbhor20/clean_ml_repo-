# Working with Dates

```python
# Converting to datetime datatype
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

# 1. Extract year

```python
date['date_year'] = date['date'].dt.year

date.head()
```

```python
date['date_month_no'] = date['date'].dt.month

date.head()
```

```python
date['date_month_name'] = date['date'].dt.month_name()

date.head()
```

```python
date['date_day'] = date['date'].dt.day

date.head()
```

```python
# day of week
date['date_dow'] = date['date'].dt.dayofweek

date.head()
```

```python
# day of week - name

date['date_dow_name'] = date['date'].dt.day_name()

date.drop(columns=['product_id','city_id','orders']).head()
```

```python
# is weekend?

date['date_is_weekend'] = np.where(date['date_dow_name'].isin(['Sunday','Saturday']),1,0)

date.drop(columns=['product_id','city_id','orders']).head()
```

# Extract week of the year

```python
date['date_week'] = date['date'].dt.isocalendar().week


date.drop(columns=['product_id','city_id','orders']).head()
```

# Extract Quarter 

```python
date['quarter'] = date['date'].dt.quarter


date.drop(columns=['product_id','city_id','orders']).head()
```

# Extract Semester

```python
# Currently, pandas does not have a built-in semester attribute in the dt accessor. However, you can calculate the semester manually based on the month.


# date['semester'] = date['date'].dt.month.apply(lambda x: 1 if x <= 6 else 2)
date['semester'] = np.where(date['quarter'].isin([1,2]), 1, 2)

date.drop(columns=['product_id','city_id','orders']).head()
```
